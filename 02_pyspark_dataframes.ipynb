```python
from pyspark.sql import SparkSession
spark= SparkSession.builder.appName('Practice').getOrCreate()
```

```python
df_pyspark=spark.read.csv("test.csv", header= True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
### Drop the columns
df_pyspark.drop('Name').show()
```

```python
## All null values record delete at single time
df_pyspark.na.drop().show()
```

```python
## How==Any 
## delete all null values records
df_pyspark.na.drop(how='any').show()
```

```python
## If only delete all null values rows
df_pyspark.na.drop(how='all').show()
```

```python
### use threshold parameter
## threshold count null values and 
## take only it which are mention
df_pyspark.na.drop(how='any', thresh=2).show()
```

```python
### use subset parameter
## subset remove null values from specific columns
df_pyspark.na.drop(how='any', subset='Salary').show()
```

```python
df_pyspark.na.drop(how='any', subset=['Age','Name']).show()
```

```python
df_pyspark.show()
```

```python
### Filing the missing values
df_pyspark.na.fill('mising_value', ['Experience','Age']).show()
```

```python
df_pyspark=spark.read.csv('Test.csv', header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
### filling missing values by using mean

from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['Age', 'Experience', 'Salary'],
    outputCols=['{}_imputed'.format(c) for c in ['Age', 'Experience', 'Salary']]
).setStrategy('mean')
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
